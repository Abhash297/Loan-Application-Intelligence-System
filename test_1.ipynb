```python
import pandas as pd

df = pd.read_csv(
    FILE_PATH,
    engine="pyarrow",
    dtype_backend="pyarrow",  # avoids Python objects
)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df = df.dropna(axis=1, how="all")  # drop columns entirely empty
```

```python
df.shape
```

```python
empty_cols = df.columns[df.isnull().sum() == len(df)]
empty_cols
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
pd.set_option("display.max_rows", 200)  # show full list
df.isnull().sum().sort_values(ascending=False)
```

```python
to_drop = df.columns[df.isnull().mean() > 0.90]
df = df.drop(columns=to_drop)
```

```python
pd.set_option("display.max_rows", 200)  # show full list
df.isnull().sum().sort_values(ascending=False)
```

```python
df = df.drop(columns=df.columns[df.isnull().mean() > 0.50])
```

```python
pd.set_option("display.max_rows", 200)  # show full list
df.isnull().sum().sort_values(ascending=False)
```

```python
very_sparse = [
    "il_util",
    "mths_since_rcnt_il",
    "all_util",
    "open_acc_6m",
    "inq_last_12m",
    "total_cu_tl",
    "open_il_12m",
    "open_rv_12m",
    "open_act_il",
    "open_il_24m",
    "inq_fi",
    "max_bal_bc",
    "total_bal_il",
    "open_rv_24m",
]

df = df.drop(columns=very_sparse, errors="ignore")
```

```python
pd.set_option("display.max_rows", 200)  # show full list
df.isnull().sum().sort_values(ascending=False)
```

```python
df.to_csv("cleaned_loan_dataset.csv", index=False)
```
